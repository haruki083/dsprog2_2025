## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること



### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5

for i in range(1, size + 1):
    print("*" * i)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 6

for i in range(size):
    print(" " * i + "*")
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 7

for i in range(size):
    print(" " * (size - 1 - i) + "*")
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 8

for i in range(size):
    for j in range(size):
        if j == i or j == size - 1 - i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 9

for i in range(size):
    spaces = size - 1 - i
    stars = 2 * i + 1
    print(" " * spaces + "*" * stars)
````
